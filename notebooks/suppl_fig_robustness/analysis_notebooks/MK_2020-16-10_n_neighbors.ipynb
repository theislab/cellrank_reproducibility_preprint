Robustness check - #neighbors
---
We check robustness with respect to the main methods parameters:

- `n_neighbors`: parameter used to construct the KNN graph which we use both for the moments (`scv.pp.moments`) and the velocity graph (`scv.tl.velocity_graph`).

To asses robustness, we use correlation between lineage probabiliteis as metric. We fix the states we are interested in to the main endpoints in this system, i.e. Alpha, Beta and Epsilon states. For each paramater combination, we record the lineage probabilities and correlate all against all. We visualize the results in heatmaps.

## Load the data

```python
adata = cr.datasets.pancreas(DATA_DIR / "pancreas" / "pancreas.h5ad")
adata.raw = adata.copy()
adata
```

### Load terminal states annotations

```python
term_states = pd.read_csv(DATA_DIR / "pancreas" / "annotations" / "terminal_states.csv", index_col=0)
term_states['0'] = term_states['0'].astype('category')
term_states
```

# Perturbing `n_neighbors`

## Non-fixed terminal states

### Deterministic mode

```python
robustness_results = {}
data = {}

for i, value in enumerate([15, 30, 50, 70, 100]):
    params, results, g_fwd, _ = run_analysis(adata, 
                                             save_figures=True,
                                             n_neighbors=value,
                                             n_metast_fwd=None,
                                             mode="deterministic",
                                             finst_fwd=['Alpha', 'Beta', 'Epsilon', 'Delta'], 
                                             return_objects=True, 
                                             show_figures=True, 
                                             cache_object=c)
    robustness_results[i] = {**params, **results}
    data[value] = g_fwd
    
with open(DATA_DIR / "benchmarking" / "robustness_analysis" / "robustness_det" / "n_neighbors.pickle", "wb") as fout:
    pickle.dump({'data': data, 'res': robustness_results}, fout)
```

### Stochastic mode

```python
robustness_results = {}
data = {}

for i, value in enumerate([15, 30, 50, 70, 100]):
    params, results, g_fwd, _ = run_analysis(adata, 
                                             save_figures=True,
                                             n_neighbors=value,
                                             n_metast_fwd=None,
                                             mode="stochastic",
                                             finst_fwd=['Alpha', 'Beta', 'Epsilon', 'Delta'], 
                                             return_objects=True, 
                                             show_figures=True, 
                                             cache_object=c)
    robustness_results[i] = {**params, **results}
    data[value] = g_fwd
    
with open(DATA_DIR / "benchmarking" / "robustness_analysis" / "robustness" / "n_neighbors.pickle", "wb") as fout:
    pickle.dump({'data': data, 'res': robustness_results}, fout)
```
